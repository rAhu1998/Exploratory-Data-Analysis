```python
## Converters

def convert_nullto_None(cell):
    if cell == "None" or cell == "n.a.":
        return None
    return cell

df = pd.read_csv("Google_Stock_Price_Train.csv",converters={
    'Date':convert_nullto_None,
    'Open':convert_nullto_None,
    'High':convert_nullto_None
})
df
```
